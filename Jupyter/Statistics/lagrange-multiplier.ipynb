```python
#
# https://www.khanacademy.org/math/multivariable-calculus/applications-of-multivariable-derivatives/lagrange-multipliers-and-constrained-optimization/v/lagrange-multiplier-example-part-2
#

import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
def R(s,h):
    return 100 * h**(2.0/3.0) * s**(1.0/3.0)
```

```python
X = np.arange(0,2000,10)
Y = np.arange(0,2000,10)

X, Y = np.meshgrid(X, Y)
Z = R(X,Y)

fig = plt.figure()
ax  = fig.gca(projection='3d')

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
plt.grid()
```
