# Solving difference equation
$$y_{n+1} = ay_n +b$$

```python
%matplotlib notebook
import numpy as np
import scipy
from scipy.signal import *
import matplotlib.pyplot as plt
```

## Using discrete samples

```python
a  = -(1.0/1.6)
b  = 5
y0 = 7.5

y    = np.zeros(10)
y[0] = y0

for i in range(1,y.size):
    y[i] = a * y[i-1] + 5
```

## More fun in the continous domain
$$y_{n+1} = ay_n +b$$
$$y_1 = ay_0 + b$$
$$y_2 = a(ay_0 +b) +b  = a^2y_0 + b(1+a)$$
$$y_3 = a(a^2y_0 + b(1-a)) + b = a^3y_0 + b(1+a+a^2)$$

$$Using: 1-a^3 = (1-a)(1+a+a^2)$$
$$y_3 = a^3y_0 + b \frac{1-a^3}{1-a}$$
$$y(x) = a^x(y_0 - \frac{b}{1-a}) + \frac{b}{1-a}$$

```python
def f(x,a,b,y0):
    return a**x * (y0 - (b/(1-a))) + (b/(1-a))
```

```python
x = np.arange(0,y.size,0.1, dtype=np.complex)

plt.figure()
plt.plot(y)
plt.plot(x,f(x, a, b, y0))
plt.grid()
```
